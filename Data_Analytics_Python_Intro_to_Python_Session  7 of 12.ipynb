Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general
Matplotlib allows you to create reproducible figures programmatically.

# **Matplotlib Object Oriented Method**

Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.


The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:



```python
#Exploring what fig and axes do
fig=plt.figure()

#This creates an empty figure that is stored in python, but as you can see there are not axes attached to it 
```

```python
#by adding axes, it creates a plot on which we can start to add our plots too
fig=plt.figure()
axes = fig.add_axes([0, 0, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
```

Now you can see that we have created an axes on our empty figure

```python
#firstly we are going to explore what the first two numbers do 
#the first two numbers represent the x and y coordinate and how  
fig=plt.figure()
axes = fig.add_axes([0, 0, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
axes = fig.add_axes([1, 0, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
axes = fig.add_axes([0, 1, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
axes = fig.add_axes([1, 1, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
```

It essesntially creates multiple plots at different x and y coordinates.
We will just keep a default of 0,0 going forward.

```python
#the next two numbers denote the height and width 
#the first two numbers represent the x and y coordinate and how  
fig=plt.figure()
axes = fig.add_axes([0, 0, 0.8, 0.8]) # xcoordinate, ycoordinate, width, height (range 0 to 1)
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'red')
axes.set_xlabel('X') # Notice the use of set_ to begin methods
axes.set_ylabel('X Squared')
axes.set_title('Squared Numbers')
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure

Before moving on to a practical example. Lets explore some of the different appearances

```python
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


# explore different marker types '+', 'o', 's', '1'
# https://matplotlib.org/stable/api/markers_api.html?highlight=marker
axes.plot(x, y, color='red', marker='o')
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


# explore marker size 
axes.plot(x, y, color='red', marker='D', markersize=10)
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


# explore linewidth
axes.plot(x, y, color='red', lw=6)
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) 


# explore colours
# you could use standard colours like red, blue, green or hex codes (https://www.color-hex.com/)
axes.plot(x, y, color='yellow', label='red line')
axes.plot(y, x, color='#009999', label = 'hex code line')
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) 


# explore linestyles
axes.plot(x, y, color='red', linestyle='--')
# More on linestyles here
#https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html?highlight=linestyles#matplotlib.lines.Line2D.lineStyles
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) 


#  explore what alpha does
axes.plot(x, y, color='red', alpha=100, label='alpha=100')
axes.plot(y, x, color='red', alpha=10, label='alpha=10')
axes.legend()
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) 

# explore what set_xlim and set_ylim do
axes.plot(x, y, color='red')
axes.set_xlim([0,5])
axes.set_ylim([0,5])
```

```python
fig = plt.figure() 

# Add set of axes to figure
axes = fig.add_axes([0, 0, 0.8, 0.8]) 
# explore what label & legend does & how does label & legend coincide
axes.plot(x, y, color='red', label='red line') 
axes.plot(y, x, color='blue', label='blue line') 
axes.legend(loc=0) # upper right = 1, upper left = 2, lower left = 3, lower_right = 4, center = 10, best location = 0
```

**How can we create multiple plots within 1 figure?**

```python
fig2, axes_2 = plt.subplots(figsize = (12,8), nrows = 2, ncols=2) #figsize represents the (width, height)
plt.tight_layout() # what does this do?

#First plot
axes_2[0,0].set_title('Plot 1')
axes_2[0,0].plot(x,y, 'red', marker='o')

#Second plot
axes_2[0,1].set_title('Plot 2')
axes_2[0,1].plot(x,x**2, 'blue', linestyle='--')

#Third plot
axes_2[1,0].set_title('Plot 3')
axes_2[1,0].plot(y,y**2, 'green')

axes_2[1,1].set_title('Plot 4')
axes_2[1,1].plot(y,x, 'yellow')
```

Exercise: Create a subplot with 2 rows and 2 columns and change the linestyles and colours for each plot.

Each plot should have a different colour, marker & linestyle.

```python
# complete exercise here
fig2, axes_2 = plt.subplots(figsize = (12,8), nrows = 2, ncols=2)
plt.tight_layout() # what does this do?

#First plot -- at row 1, column 1
axes_2[0,0].set_title('Plot 1')
axes_2[0,0].plot(x,y, 'red', marker='o', linestyle='--')

#Second plot -- at row 1, column 2
axes_2[0,1].set_title('Plot 2')
axes_2[0,1].plot(x,x**2, 'blue', linestyle='-.', marker='s')

#Third plot -- at row 2, column 1
axes_2[1,0].set_title('Plot 3')
axes_2[1,0].plot(y,y**2, 'green', linestyle=':')

#fourth plot -- at row 2, column 2
axes_2[1,1].set_title('Plot 4')
axes_2[1,1].plot(y,x,'orange')
```

How do i create a plot with multiple lines within the same axes?

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.plot(x,x*2, label='x times 2')
ax.plot(x,x+5, label='x+5')
ax.set_xlabel('x')
ax.legend(loc=10)
```
